```python
dataRepo='https://github.com/Controles-ICDE-Ochoa/PC3_2024.2/raw/d516b1cb4f3fc8664a280227547618747279ffd1/appendData'
levels=range(1,5)
fileNames=['/level'+str(level)+'.xlsx' for level in levels]
alltheLinks=[dataRepo+fn for fn in fileNames]
alltheLinks
```

***LIMPIEZA***

```python
#Primer link
import pandas as pd
link=alltheLinks[0]
data = pd.read_excel(link)
data.drop(index=[0, 1, 2], inplace=True)
nuevos_titulos=list(data.iloc[0])
data.columns = nuevos_titulos
data.drop(index=3,inplace=True)
data_nan=data.dropna(axis=1, how='all')
data_cleaned = data_nan.loc[:, data_nan.columns.notna()].copy()
data_cleaned.reset_index(drop=True, inplace=True)
data_cleaned.drop(index=[0, 1], inplace=True)
data_cleaned1 = data_cleaned.dropna(subset=['HDI rank'])
data_cleaned2 = data_cleaned1.replace(',', '.', regex=True)
data_cleaned2.iloc[:, 2:] = data_cleaned2.iloc[:, 2:].astype(float)
data_cleaned2.reset_index(drop=True, inplace=True)
data_cleaned2 = data_cleaned2.loc[:, ~data_cleaned2.columns.duplicated()]
data_cleaned2
```

```python
#Segundo link
link=alltheLinks[1]
data = pd.read_excel(link)
data.drop(index=[0, 1], inplace=True)
data.drop(index=3,inplace=True)
data_nan=data.dropna(axis=1, how='all')
data_cleaned = data_nan.dropna(axis=1)
data_cleaned.reset_index(drop=True, inplace=True)
data2_cleaned2 = data_cleaned.replace(',', '.', regex=True)
data2_cleaned2.iloc[:, 2:] = data2_cleaned2.iloc[:, 2:].astype(float)
data2_cleaned2
```

```python
#Tercer link
link=alltheLinks[2]
data = pd.read_excel(link)
data.drop(index=[0, 1], inplace=True)
data.drop(index=3,inplace=True)
data_nan=data.dropna(axis=1, how='all')
data_cleaned = data_nan.dropna(axis=1)
data_cleaned.reset_index(drop=True, inplace=True)
data3_cleaned2 = data_cleaned.replace(',', '.', regex=True)
data2_cleaned2.iloc[:, 2:] = data2_cleaned2.iloc[:, 2:].astype(float)
data3_cleaned2
```

```python
#cuarto link
link=alltheLinks[-1]
data = pd.read_excel(link)
data_nan=data.dropna(axis=1, how='all')
data_cleaned = data_nan.dropna(axis=1)
data_cleaned.reset_index(drop=True, inplace=True)
data4_cleaned2 = data_cleaned.replace(',', '.', regex=True)
data4_cleaned2 = data4_cleaned2.drop(columns=[data4_cleaned2.columns[2]])
data4_cleaned2.iloc[:, 2:] = data4_cleaned2.iloc[:, 2:].astype(float)
data4_cleaned2
```

***PARTE I***

```python
import pandas as pd
allDFs=[data_cleaned2, data2_cleaned2, data3_cleaned2, data4_cleaned2]
allColumnNames=[]
for df in allDFs:
    allColumnNames.append(set(df.columns))
[len(cols) for cols in allColumnNames]
```

```python
commonColumns=set.intersection(*allColumnNames)
len(commonColumns),commonColumns
```

```python
set.union(*allColumnNames)-commonColumns
```

```python
allDFs_sameNames=[df.loc[:,list(commonColumns)] for df in allDFs]
allDFsConcat=pd.concat(allDFs_sameNames)
allDFsConcat.reset_index(drop=True, inplace=True)
allDFsConcat.columns = allDFsConcat.columns.str.replace(' ', '', regex=True)
allDFsConcat
```
